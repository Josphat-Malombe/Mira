```python
import torch
import transformers
import bitsandbytes
from transformers import AutoTokenizer, AutoModelForCausalLM, Pipeline, BitsAndBytesConfig
from peft import LoraConfig, get_peft_model
from datasets import load_dataset
from google.colab import drive
from transformers import TrainingArguments, Trainer, DataCollatorForLanguageModeling

```

```python
device="cuda" if torch.cuda.is_available() else "cpu"
```

```python
model_id="Qwen/Qwen3-1.7B"
```

```python
bnb_config=BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)
```

```python
model=AutoModelForCausalLM.from_pretrained(model_id, quantization_config=bnb_config,device_map="auto")
tokenizer=AutoTokenizer.from_pretrained(model_id)
```

```python
lora_config=LoraConfig(
    r=8,
    lora_alpha=16,
    target_modules=["q_proj","k_proj"],
    lora_dropout=0.10,
    bias="none",
    task_type="CAUSAL_LM"
)
```

```python
model=get_peft_model(model,lora_config)
model.print_trainable_parameters()
```

```python
drive.mount('/content/drive')
path="/content/drive/MyDrive/mira_dataset2.json"
mira_dataset=load_dataset("json",data_files=path)
```

```python
train_val_split=mira_dataset["train"].train_test_split(test_size=0.1)
train_data=train_val_split["train"]
val_data=train_val_split["test"]

print(f"Train length: {len(train_data)}")
print(f"Validation length: {len(val_data)}")
```

```python
train_data=train_data.map(tokenize_function,batched=True)
val_data=val_data.map(tokenize_function,batched=True)

train_data.set_format("torch")
val_data.set_format("torch")
```

```python
data_collator=DataCollatorForLanguageModeling(tokenizer=tokenizer,mlm=False)
```

```python
training_args=TrainingArguments(
    output_dir="./Mira_Model",
    overwrite_output_dir=True,
    num_train_epochs=3,
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    save_steps=500,
    save_total_limit=2,
    save_strategy="steps",
    #evaluation_strategy="epoch",
    logging_steps=50,
    logging_dir="./Mira_Logs",
    warmup_steps=100,
    weight_decay=0.01,
    fp16=True,
    push_to_hub=False,
    #load_best_model_at_end=True

)
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM

model_path = "/content/drive/MyDrive/mira_model"
try:
    tokenizer = AutoTokenizer.from_pretrained(model_path)
    model = AutoModelForCausalLM.from_pretrained(model_path).to("cuda" if torch.cuda.is_available() else "cpu")
except Exception as e:
    raise Exception(f"Failed to load model or tokenizer: {str(e)}")

if tokenizer.eos_token_id is None:
    tokenizer.eos_token = "<|endoftext|>"
    tokenizer.eos_token_id = tokenizer.encode("<|endoftext|>")[0]


SYSTEM_PROMPT = "You are Mira, a helpful AI assistant. Respond concisely and accurately.\n"


class ChatSession:
    def __init__(self):
        self.history = SYSTEM_PROMPT
        self.max_history_tokens = 512

    def append_to_history(self, user_input, ai_response):
        """Append user input and AI response to history, trimming if necessary."""
        self.history += f"User: {user_input}\n\nAI {ai_response}\n"

        tokenized_history = tokenizer(self.history, return_tensors="pt")
        if tokenized_history.input_ids.shape[1] > self.max_history_tokens:
            recent_tokens = tokenized_history.input_ids[0, -(self.max_history_tokens - len(tokenizer(SYSTEM_PROMPT).input_ids[0])):]
            self.history = SYSTEM_PROMPT + tokenizer.decode(recent_tokens, skip_special_tokens=True)

    def get_history(self):
        """Return current chat history."""
        return self.history

    def reset(self):
        """Reset chat history to initial state."""
        self.history = SYSTEM_PROMPT

def chat_with_mira(user_input, session=None, max_new_tokens=50, diagnostic=False):
    """
    Generate a response from the Mira model for the given user input.

    Args:
        user_input (str): The user's input text.
        session (ChatSession, optional): Session object to maintain chat history.
        max_new_tokens (int): Maximum number of new tokens to generate.
        diagnostic (bool): If True, print diagnostic information.

    Returns:
        str: The model's response.
    """
    if not user_input or not isinstance(user_input, str):
        return "Error: Invalid or empty input."


    if session is None:
        session = ChatSession()

    try:
        # Update chat history with user input
        current_history = session.get_history()
        temp_history = current_history + f"User: {user_input}\n\nAI"

        if diagnostic:
            print(f"[DIAG] Input history sent to tokenizer:\n{temp_history}")

        # Tokenize input
        inputs = tokenizer(temp_history, return_tensors="pt").to(model.device)

        # Generate response
        outputs = model.generate(
            **inputs,
            max_new_tokens=max_new_tokens,
            pad_token_id=tokenizer.eos_token_id,
            do_sample=True,
            top_p=0.95,
            temperature=0.7,

        )

        response = tokenizer.decode(outputs[0], skip_special_tokens=True)


        new_response = response[len(temp_history):].strip()

        for stop_marker in ["User:", tokenizer.eos_token, "\n\nUser:"]:
            if stop_marker in new_response:
                new_response = new_response.split(stop_marker)[0].strip()

        if diagnostic:
            print(f"[DIAG] Full decoded response:\n{response}")
            print(f"[DIAG] Extracted new response:\n{new_response}")

        # Update session history with AI response
        session.append_to_history(user_input, new_response)

        return new_response

    except Exception as e:
        return f"Error generating response: {str(e)}"


if __name__ == "__main__":
    session = ChatSession()
    session.reset()
    print("Starting chat with Mira. Type 'exit' to quit.")
    while True:
        user_input = input("You: ")
        if user_input.lower() == "exit":
            break
        response = chat_with_mira(user_input, session=session, max_new_tokens=50, diagnostic=False)
        print(f"Mira: {response}")
```
